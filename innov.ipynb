```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import optuna
from sklearn.preprocessing import StandardScaler

import xgboost as xgb
import numpy as np
from sklearn.metrics import mean_squared_error

def objective(trial, X_train, y_train, X_test, y_test):
    """Функция для оптимизации гиперпараметров XGBoost с помощью Optuna."""
    
    params = {
        'objective': 'reg:squarederror',  # Указываем 'reg:squarederror' для регрессии
        'eval_metric': 'mae',            # Метрика оценки
        'tree_method': 'hist',            # Для ускорения
        'n_estimators': trial.suggest_int('n_estimators', 60, 75),
        'max_depth': trial.suggest_int('max_depth', 60, 75),
        'learning_rate': trial.suggest_float('learning_rate', 1e-1, 0.15, log=True),
        'subsample': trial.suggest_float('subsample', 0.6, 0.8),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.8, 1.0),
        'min_child_weight': trial.suggest_int('min_child_weight', 7, 12),
        'gamma': trial.suggest_float('gamma', 1e-9, 1e-7),
    }

    model = xgb.XGBRegressor(**params)
    model.fit(X_train, y_train,
              eval_set=[(X_test, y_test)],  # Передаем тестовые данные
              verbose=False)

    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    return rmse
```

{'n_estimators': 67, 'max_depth': 64, 'learning_rate': 0.09111390813822273, 'subsample': 0.7221429904553741, 'colsample_bytree': 0.96109561986183, 'min_child_weight': 9, 'gamma': 2.4912900684746213e-08}

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1])


# Оптимизируем гиперпараметры XGBoost с помощью Optuna
study = optuna.create_study(direction='minimize')
study.optimize(lambda trial: objective(trial, X_train, y_train, X_test, y_test), n_trials=100)

print("Лучшие гиперпараметры:", study.best_params)
print("Лучшее MAE:", study.best_value)
```
